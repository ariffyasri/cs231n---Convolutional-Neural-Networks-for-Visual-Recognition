```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sin and  Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```
